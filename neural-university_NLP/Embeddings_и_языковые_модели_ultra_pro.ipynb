```python
# –¢–æ–∫–µ–Ω–∏–∑–∞—Ç–æ—Ä –¥–ª—è –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —Ç–µ–∫—Å—Ç–æ–≤ –≤ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏
from tensorflow.keras.preprocessing.text import Tokenizer
# –§—É–Ω–∫—Ü–∏–∏-—É—Ç–∏–ª–∏—Ç—ã –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏
from tensorflow.keras import utils,regularizers
# –†–∞–±–æ—Ç–∞ —Å –º–∞—Å—Å–∏–≤–∞–º–∏ –¥–∞–Ω–Ω—ã—Ö
import numpy as np 
# –ö–ª–∞—Å—Å –¥–ª—è –∫–æ–Ω—Å—Ç—Ä—É–∏—Ä–æ–≤–∞–Ω–∏—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ–π –º–æ–¥–µ–ª–∏ –Ω–µ–π—Ä–æ–Ω–Ω–æ–π —Å–µ—Ç–∏
from tensorflow.keras.models import Sequential
# –û—Å–Ω–æ–≤–Ω—ã–µ —Å–ª–æ–∏
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, Activation ,Embedding,SpatialDropout1D,Flatten,Input,LSTM,Conv1D
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.optimizers import Adam
# –§—É–Ω–∫—Ü–∏–∏ –æ–ø–µ—Ä–∞—Ü–∏–æ–Ω–Ω–æ–π —Å–∏—Å—Ç–µ–º—ã
import os
# –†–µ–≥—É–ª—è—Ä–Ω—ã–µ –≤—ã—Ä–∞–∂–µ–Ω–∏—è
import re
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
import zipfile
import xml.etree.ElementTree

WORD_NAMESPACE = '{http://schemas.openxmlformats.org/wordprocessingml/2006/main}'
PARA = WORD_NAMESPACE + 'p'
TEXT = WORD_NAMESPACE + 't'
TABLE = WORD_NAMESPACE + 'tbl'
ROW = WORD_NAMESPACE + 'tr'
CELL = WORD_NAMESPACE + 'tc'

with zipfile.ZipFile('Tesla 1.docx') as docx:
    tree = xml.etree.ElementTree.XML(docx.read('word/document.xml'))

for table in tree.iter(TABLE):
    for row in table.iter(ROW):
        for cell in row.iter(CELL):
            print(''.join(node.text for node in cell.iter(TEXT)))
```

```python
!pip install docx2txt
import docx2txt
text1 = docx2txt.process("Tesla 1.docx")
text2 = docx2txt.process("Tesla 2.docx")
text1.replace('\xa0','')
text2.replace('\xa0','')
text1.replace('.','')
text2.replace('.','')
```

```python
text1_list=text1.split('\n')
text1_list=[item for item in text1_list if item!='']
print(len(text1_list))
text2_list=text2.split('\n')
text2_list=[item for item in text2_list if item!='']
print(len(text2_list))
```

```python
print(text1_list[:10])
```

–ü–∞—Ä—Å–∏–Ω–≥ –¥–∞–Ω–Ω—ã—Ö, —Ä–∞–∑–±–∏–≤–∫–∞ –Ω–∞ –∫–ª–∞—Å—Å—ã.

```python
# y1 -1  0  1  y2  + 1,0 -0,1 +- 1,1  y3 +1,0,0 - 0,0,1  +- 0,1,0
x_data_text=[]
y_data1=[]
y_data2=[]
y_data3=[]
for item in text1_list:
  if(item[:2]=='+ '):
    x_data_text.append(item[2:])
    y_data1.append(np.array(1))
    y_data2.append(np.array((1,0)))
    y_data3.append(np.array((1,0,0)))
  elif(item[:2]=='- '):
    x_data_text.append(item[2:])
    y_data1.append(np.array(-1))
    y_data2.append(np.array((0,1)))
    y_data3.append(np.array((0,0,1)))
  elif(item[:2]=='+-' or item[:2]=='-+'):
    text=item[2:]
    if(text[0]==' '):
      text=text[1:]
    x_data_text.append(text)
    y_data1.append(np.array(0))
    y_data2.append(np.array((1,1)))
    y_data3.append(np.array((0,1,0)))
for item in text2_list:
  if(item[:2]=='+ '):
    x_data_text.append(item[2:])
    y_data1.append(np.array(1))
    y_data2.append(np.array((1,0)))
    y_data3.append(np.array((1,0,0)))
  elif(item[:2]=='- '):
    x_data_text.append(item[2:])
    y_data1.append(np.array(-1))
    y_data2.append(np.array((0,1)))
    y_data3.append(np.array((0,0,1)))
  elif(item[:2]=='+-' or item[:2]=='-+'):
    text=item[2:]
    if(text[0]==' '):
      text=text[1:]
    x_data_text.append(text)
    y_data1.append(np.array(0))
    y_data2.append(np.array((1,1)))
    y_data3.append(np.array((0,1,0)))
y_data1=np.array(y_data1)
y_data1=y_data1.reshape((-1,1))
y_data2=np.array(y_data2)
y_data3=np.array(y_data3)
print(len(x_data_text))
print(y_data1.shape,y_data2.shape,y_data3.shape)
```

```python
print(x_data_text[:5])
```

```python
VOCAB_SIZE=20000
# –¢–æ–∫–µ–Ω–∞–π–∑–µ—Ä –∏–∑ Keras –¥–ª—è —Ä–∞–∑–±–∏–µ–Ω–∏—è —Ç–µ–∫—Å—Ç–∞ –∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è —á–∞—Å—Ç–æ—Ç–Ω–æ–≥–æ —Å–ª–æ–≤–∞—Ä—è
tokenizer = Tokenizer(num_words=VOCAB_SIZE, filters='!"#$%&()*+,-‚Äì‚Äî./‚Ä¶:;<=>?@[\\]^_`{|}~¬´¬ª\t\n\xa0\ufeff', 
                          lower=True, split=' ', oov_token='–Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–æ–µ_—Å–ª–æ–≤–æ', char_level=False)
```

```python
# –ü–æ–ª—É—á–∞–µ–º —Å–ª–æ–≤–∞—Ä—å —á–∞—Å—Ç–æ—Ç–Ω–æ—Å—Ç–∏ 
tokenizer.fit_on_texts(x_data_text)
```

```python
seq_train = tokenizer.texts_to_sequences(x_data_text)
```

```python
print(y_data1[:5])
print(y_data2[:5])
print(y_data3[:5])
```

```python
#–û–±—Ä—É–±–∏–º –æ—Ç–∑—ã–≤—ã –¥–æ 17 —Å–ª–æ–≤ –∏ –¥–æ–ø–æ–ª–Ω–∏–º –Ω—É–ª—è–º–∏ –µ—Å–ª–∏ –Ω–µ —Ö–≤–∞—Ç–∞–µ—Ç
SequencyWords=17
train_data = pad_sequences(seq_train, maxlen=SequencyWords, padding='post', truncating='post') 
```

```python
train_data_balance=train_data.copy()
y_data1_balance=y_data1.copy()
y_data2_balance=y_data2.copy()
y_data3_balance=y_data3.copy()
add_c0=2171-1152
add_c1=2171-1568
train_data_balance=np.vstack((train_data_balance,train_data[y_data1[:,0]==-1][:add_c0,:]))
y_data1_balance=np.vstack((y_data1_balance,y_data1[y_data1[:,0]==-1][:add_c0].reshape(-1,1)))
y_data2_balance=np.vstack((y_data2_balance,y_data2[y_data1[:,0]==-1][:add_c0]))
y_data3_balance=np.vstack((y_data3_balance,y_data3[y_data1[:,0]==-1][:add_c0]))
train_data_balance=np.vstack((train_data_balance,train_data[y_data1[:,0]==0][:add_c1,:]))
y_data1_balance=np.vstack((y_data1_balance,y_data1[y_data1==0][:add_c1].reshape(-1,1)))
y_data2_balance=np.vstack((y_data2_balance,y_data2[y_data1[:,0]==0][:add_c1]))
y_data3_balance=np.vstack((y_data3_balance,y_data3[y_data1[:,0]==0][:add_c1]))
print("class0=",np.shape(y_data1_balance[y_data1_balance==-1]))
print("class1=",np.shape(y_data1_balance[y_data1_balance==0]))
print("class2=",np.shape(y_data1_balance[y_data1_balance==1]))
```

```python
l2Conv=0.03
l2Dense=0.03

input = Input((SequencyWords))
x = Embedding(VOCAB_SIZE, 20)(input)
x = SpatialDropout1D(0.2)(x)
x = Conv1D(8,3, activation="relu",kernel_regularizer=regularizers.l2(l2Conv))(x)#,bias_regularizer=regularizers.l2(l2Conv)
x = Flatten()(x)
#x=Dense(100, activation="relu")(x)
#x = BatchNormalization()(x)
x = Dropout(0.5)(x)
x = Dense(30, activation="relu",kernel_regularizer=regularizers.l2(l2Conv))(x)
#x = BatchNormalization()(x)
#x = Dropout(0.5)(x)
#x = Dense(33, activation="relu",kernel_regularizer=regularizers.l2(l2Conv))(x)
x = BatchNormalization()(x)
x = Dropout(0.1)(x)
y1= Dense(1,activation='tanh')(x)
y2= Dense(2,activation='sigmoid')(x)
y3= Dense(3,activation='softmax')(x)

model=Model(input,[y1,y2,y3])
model.compile(loss='mse',metrics=['accuracy'], optimizer=Adam(learning_rate=1e-4))
```
